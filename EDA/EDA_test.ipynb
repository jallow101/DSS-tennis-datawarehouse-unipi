```python
#sns plot missing values
sns.heatmap(pd.isnull(), cbar=False, cmap='viridis')
#sns plot missing values
plt.title('Missing Values in Dataset')
plt.show()
```

```python
#calculate percentage of missing values
missing_percentage = pd.isnull().sum() / len(pd) * 100
missing_percentage = missing_percentage[missing_percentage > 0]
missing_percentage = missing_percentage.sort_values(ascending=False)
missing_percentage = missing_percentage.reset_index()
missing_percentage.columns = ['Column Name', 'Missing Percentage']
#plot missing values
plt.figure(figsize=(10, 6))
sns.barplot(x='Missing Percentage', y='Column Name', data=missing_percentage, palette='viridis')
```

```python
missing_percentage
```

```python
#drop columns with more than 59% missing values in missing_percentage, 1- 21
for col in missing_percentage['Column Name']:
    if missing_percentage[missing_percentage['Column Name'] == col]['Missing Percentage'].values[0] > 59:
        pd.drop(columns=[col], inplace=True)
```

```python
pd.drop(columns=["winner_seed"], inplace=True)
```
